```python
import pandas as pd
import numpy as np
import csv
import json
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Social_Progress_Index'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
soup.find('tr')
```

```python
world_titles = soup.find('tr')
world_table_titles = [title.text.strip() for title in world_titles]
```

```python
world_table_titles
```

```python
data_without_spaces = list(filter(lambda x: x.strip(), world_table_titles))

print(data_without_spaces)
```

```python
dftable = pd.DataFrame(columns = data_without_spaces)

dftable
```

```python
table=soup.find('tbody')
table
```

```python
rows = table.find_all('tr')[1:]
rows
```

```python
for row in rows:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(dftable)
    dftable.loc[length] = individual_row_data
```

```python
rows
```

```python
dftable.insert(0, 'New_Column', None)
dftable
```

```python
dftable.to_csv('socialprogress.csv', index=False)
```
