```python
import pandas as pd
import numpy as np
import csv
import json
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.theglobaleconomy.com/rankings/wb_political_stability/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
table=soup.find_all('tr')[0]
table
```

```python
world_table_titles = [title.text.strip() for title in table]
world_table_titles
```

```python
data_without_spaces = list(filter(lambda x: x.strip(), world_table_titles))

print(data_without_spaces)
```

```python
dftable = pd.DataFrame(columns = data_without_spaces)

dftable
```

```python
table=soup.find_all('tbody')[0]
table
```

```python
rows = table.find_all('tr')
rows
```

```python
for row in rows:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(dftable)
    dftable.loc[length] = individual_row_data
```

```python
dftable
```

```python
dftable.to_csv('output.csv', index=False)
```
