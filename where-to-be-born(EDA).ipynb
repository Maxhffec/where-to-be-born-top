```python
df['gpd per capita'] = pd.to_numeric(df['gpd per capita'], errors='coerce')
df['Gender Development Index (2021)'] = pd.to_numeric(df['Gender Development Index (2021)'], errors='coerce')
df.fillna(0, inplace=True)
```

```python
df['total_score'] = df['Gender Development Index (2021)']*9.69+ df['HUMAN CAPITAL INDEX 2020']*11.3636 + \
                    ((df['Political stability, 2022']+2.550000)*1.25) + df['BasicHumanNeeds']*0.1614 + df['Opportunity']*0.16635 +  \
                    df['Water andSanitation']*0.1011+ df['PersonalSafety']*0.1803 + df['Accessto BasicKnowledge']*0.15045+df['EnvironmentalQuality']*0.1176+ \
                    df['PersonalRights']*0.15285+df['HealthandWellness']*0.2 + df['Inclusiveness']*0.1086+df['Access toAdvancedEducation']*0.1719+ \
                    df['Social support']*15.26-df['Living index']*0.0811+ df['gpd per capita']*0.000043
```
